```python
#task 4 file

#task 4.2 image swapping
import cv2 as cv

image1 = cv.imread("imageA.png")
image2 = cv.imread("imageB.png")


crop1 = image1[100: 400, 50:400]
crop2 = image2[100: 400, 50:400]

cv2.imwrite("swap.png", crop1)
swap = cv.imread("swap.png")
  
image1[100: 400, 50:400] = crop2
image2[100: 400, 50:400] = swap

cv.imshow('window', image1)
cv.waitKey(0)
cv.destroyAllWindows()
  
cv.imshow('window', image2)
cv.waitKey(0)
cv.destroyAllWindows()
  
```

```python
# Task 4.3

import cv2
import numpy as np


image1=cv2.imread("cat1.jpg")
image2=cv2.imread("cat2.jpg")

#to make collage, images should be of same size

image1=cv2.resize(image1,(300,400))
image2=cv2.resize(image2,(300,400))

# we will use numpy stack function
Horizontal1 = np.hstack([image1,image2])

cv2.imshow("Final Collage", Horizontal1 )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#task 4.1 House 

import cv2  
import numpy as np 
width=1000
height=700 
blank_image = np.zeros((height,width,3), np.uint8) 
blank_image[:,:]=(0,0,0) 

cv2.rectangle(blank_image, (100, 350), (300, 600), (0, 165, 255), -2)
cv2.rectangle(blank_image, (305, 350), (700, 600), (122,160,255), -2)

#window
cv2.rectangle(blank_image, (350, 400), (650, 550), (255, 255, 255), 5)

cv2.rectangle(blank_image, (350, 400), (495, 475), (255, 255, 255), 3)#1
cv2.rectangle(blank_image, (500, 400), (650, 475), (255, 255, 255), 3)#2

cv2.rectangle(blank_image, (350, 480), (495, 550), (255, 255, 255), 3)#3
cv2.rectangle(blank_image, (500, 480), (650, 550), (255, 255, 255), 3)#4



#door
cv2.rectangle(blank_image, (150, 400), (250, 595), (255, 255, 255), 5)
cv2.line(blank_image, (200, 402), (200, 588), (255, 255, 255), 5)

#roof
cv2.line(blank_image, (200, 200), (100, 350), (160,158,95), 5)
cv2.line(blank_image, (200, 200), (305, 350), (160,158,95), 5)

cv2.line(blank_image, (600, 200), (700, 350), (180,130,70), 5)
cv2.line(blank_image, (200, 200), (600, 200), (180,130,70), 5)


cv2.circle(blank_image, (200, 300), 40, (255, 255, 255), -1)


cv2.imshow('hi' , blank_image) 
cv2.waitKey() 
cv2.destroyAllWindows()
```
